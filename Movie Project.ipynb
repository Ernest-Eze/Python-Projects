```python
# Import pandas and read data & data types
import pandas as pd
df=pd.read_csv(r'C:\Users\winds\Desktop\Python Proj\movies.csv')
df.dtypes
```

```python
#A glipse of the data
df.head
```

```python
#Check for Missing values in the data sets
import numpy as np
for col in df.columns:
    Percent_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,Percent_missing))
```

```python
#Change Data Type (Float to Int)

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['budget'].astype('int64')
```

```python
#create a new column "yearcorrect" to address discrepancy in release year vs year
df['yearcorrect'] = df['released'].astype(str).str[:4]
df.head()
```

```python
#Order by gross revenue
df.sort_values(by=['gross'], inplace=False, ascending=False)
df.head()
```

```python
#set option to view all rows
#pd.set_option('display.max_rows', None)
```

```python
#Drop any duplicates (No duplicates from the data)
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
#Budget Correlation
#Scatter plot ( budget vs gross revenue)
import matplotlib.pyplot as plt

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show
```

```python
#Plotting Budget versus Gross using seaborn
import seaborn as sns
sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color":"blue"})
```

```python
#Looking at correlation (Note the high correlation between budget and gross)
df.corr(method='pearson') #Other methods are "kendall, "spearman'
```

```python
#Creating the above correlation in a colourful matrix
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
#Numerizing strings (converting all strings to Integers)
df_numerized = df

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype =="object"):
        df_numerized[col_name] = df_numerized[col_name].astype("category")
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
```

```python
#showing all fields as a correlation in a colourful matrix
correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
#Unstacking (Comparing correlation of 1 item against all others in a stack)
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
#sorted (like the matrix above but in a sort of linear way)
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
#Showing only high correlations greater than 0.5
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
